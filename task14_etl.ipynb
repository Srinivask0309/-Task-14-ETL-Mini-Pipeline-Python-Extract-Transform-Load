```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv('raw/retail_sales_raw.csv')
df.head()
```

```python
df.columns = df.columns.str.lower()
df['margin']=df['profit']/df['sales']
```

```python
conn=sqlite3.connect('output/database.sqlite')
df.to_sql('orders',conn,if_exists='replace',index=False)
conn.close()
```
